```python
 # value, wage and release clause conversion
def value_and_wage_conversion(Value):
  if isinstance(Value,str):
    out = Value.replace('€','')
    print('out',out)
    if 'M' in out :
      out = float(out.replace('M',''))*1000000
      print('m',out)
    elif 'K' in Value :
      out = float(out.replace('K',''))*1000
      print('k',out)
    return float(out)

football['Value'] = football['Value'].apply(lambda x: value_and_wage_conversion(x))
print(football['Value'])
football['Wage']= football['Wage'].apply(lambda x: value_and_wage_conversion(x))
print(football['Wage'])
football['Release Clause'] = football['Release Clause'].apply(lambda x: value_and_wage_conversion(x))
print(football['Release Clause'])
football['Release Clause'].fillna(football['Release Clause'].mean(),inplace=True)
print(football['Release Clause'])
```

```python
## Categorial Columns
football['Club'].fillna('No Club', inplace = True)
print(football['Club'])
football['Position'].fillna('unkown', inplace = True)
print(football['Position'])
```

```python
## Joined Column

def Clean_data(x):
  l=[]
  if isinstance(x, str):
    l = x.split(", ")
    print("l :",l)
    x=l[-1]
    print("x :",x)
  return(x)
football['Joined'] = football['Joined'].replace(np.nan,0)
print("football['Joined']",football['Joined'])
football['Joined'] = football['Joined'].apply(Clean_data).astype('str')
print("football['Joined']",football['Joined'])
```

```python
# Contract Valid Until Column

import datetime
football['Contract Valid Until'] = pd.to_datetime(football['Contract Valid Until'])
print("football['Contract Valid Until']",football['Contract Valid Until'])
football['year'] = pd.DatetimeIndex(football['Contract Valid Until']).year.astype(str)
print("football['year']",football['year'])
football['Contract Valid Until'].fillna('Not available',inplace=True)
print("football['Contract Valid Until']",football['Contract Valid Until'])
```

```python
# Height Column

def clean_heaight(x):
  l=[]
  s=1
  if isinstance(x,str):
    l = x.split("'")
    print("l :",l)
    i = int(l[0])
    print("i :",i)
    j = int(l[1])
    print("j :",j)
    s = ((i*12)+j)/12
    print("s :",s)
  return (s)

football['Height'] = football['Height'].apply(clean_heaight)
print("football['Height'] :",football['Height'])
football['Height'].fillna((football['Height'].mean()),inplace = True)
print("football['Height'] :",football['Height'])
```

```python
# Weight Column

def clean_weight(x):
  if isinstance(x,str):
    return(x.replace('lbs',''))
  return(x)

football['Weight'] = football['Weight'].apply(clean_weight).astype('float')
print("football['Weight'] :",football['Weight'])
football['Weight'].fillna((football['Weight'].mean()),inplace=True)
print("football['Weight'] :",football['Weight'])
```

```python
# Skill Column

## Function to convert skill rating at each position

def skillconverter(val):
  if type(val) == str:
    s1 = val[0:2]
    s2 = val[-1]
    val = int(s1) + int(s2)
    return val 
  else:
    return val

skill_column = ['LS','ST','RS','LW','LF','CF','RF','RW','LAM','CAM','RAM','LM',
                'LCM','CM','RCM','RM','LWB','LDM','CDM','RDM','RWB','LB','LCB','CB','RCB','RB']
for col in skill_column:
  football[col] = football[col].apply(skillconverter)
  football[col].fillna(0.0,inplace=True)
```

```python
## Number of Players for each categorical variables

def pie_count(data, field='Nationality',percent_limit=0.5,title='Number of players by'):
  title += field
  #data[field] = data[field].fillna('NA)
  data = data[field].value_counts().to_frame()
  total = data[field].sum()
  data['percentage'] = 100 * data[field]/total

  percent_limit = percent_limit
  otherdata = data[data['percentage'] < percent_limit]
  others = otherdata['percentage'].sum()
  maindata = data[data['percentage'] >= percent_limit]

  data = maindata
  other_label = "others(<"  + str(percent_limit) + "% each)"
  data.loc[other_label] = pd.Series({field:otherdata[field].sum()})
  
  labels = data.index.tolist()
  datavals = data[field].tolist()
  trace = go.Pie(labels = labels, values = datavals)
  layout = go.Layout(title = title, height=700)
  fig = go.Figure(data=[trace],layout=layout)
  fig.show()
pie_count(football,'Nationality')
pie_count(football,'Club')
pie_count(football,'Preferred Foot')
pie_count(football,'Work Rate', 0.1)
pie_count(football,'Body Type', 0.1)
pie_count(football,'Position', 0.1)
```

```python
## Top Features for every Playing Position

player_features = ['Crossing','Finishing','HeadingAccuracy','ShortPassing','Volleys','Dribbling',
                   'Curve','FKAccuracy','LongPassing','BallControl','Acceleration','SprintSpeed',
                   'Agility','Reactions','Balance','ShotPower','Jumping','Stamina','Strength',
                   'LongShots','Aggression','Interceptions','Positioning','Vision','Penalties',
                   'Composure','Marking','StandingTackle','SlidingTackle','GKDiving',
                   'GKHandling','GKKicking','GKPositioning','GKReflexes']
df_position = pd.DataFrame()
for position_name, features in football.groupby(football['Position'])[player_features].mean().iterrows():
  top_features = dict(features.nlargest(5))
  df_position[position_name] = tuple(top_features)
df_position.head()
```

```python
## Dream Team

position = []
player = []
club_1 = []
for col in df_position.columns:
  tmp_df = pd.DataFrame()
  l = [df_position[col].values]
  l = l[0]
  l = list(l)
  l.append('Name')
  tmp_df = pd.DataFrame.copy(football[football['Position']==col][:])
  tmp_df['mean'] = np.mean(tmp_df.iloc[:, :-1],axis=1)
  name = tmp_df['Name'][tmp_df['mean']==tmp_df['mean'].max()].values[:]
  club = football['Club'][football['Name']==str(name)].values[:1]
  position.append(col)
  player.append(name)
  club_1.append(club)

gk = ['GK']
forward = ['LS','ST','RS','LF','CF','RF']
midfeilder = ['LW','RW','LAM','CAM','RAM','LM','RCM','RM','LDM','CDM','RDM','LCM','CM']
defenders = ['LWB','RWB','LB','LCB','CB']

print('GoalKeeper : ')
for p , n , c in zip(position,player,club_1):
  if p in gk:
    print('{}[Club : {},Position : {}]'.format(n,c,p))
print('\nFORWARD :')
for p , n , c in zip(position,player,club_1):
  if p in forward:
    print('{}[Club : {},Position : {}]'.format(n,c,p))
print('\nMIDFEILDER :')
for p , n , c in zip(position,player,club_1):
  if p in midfeilder:
    print('{}[Club : {},Position : {}]'.format(n,c,p))
print('\nDEFENDER :')
for p , n , c in zip(position,player,club_1):
  if p in defenders:
    print('{}[Club : {},Position : {}]'.format(n,c,p))
print('\nFORWARD :')
```
